# Mission 3

Crée une fonction qui prend en paramètre un dataframe, et affiche le taux de remplissage de chacune de ces colonnes sous forme de pourcentage. C'est à dire :
- colonne 1 : 73%
- colonne 2 : 99%

```
def notmissing(df):
  df1 = (round((df.notnull().sum() * 100/ len(df)),2).sort_values(ascending=False))
  return df1
df1 = notmissing(df)  #to regester the result of function into a dataframe  
```

```
df1.head(60) #Top 40 columns with fillrate in percentage
```
